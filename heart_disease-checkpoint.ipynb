```python
clean=df.dropna()
clean.isnull().sum() #ends here by checking if extra data is prsent, if yes, it gets dropped
```

```python
import matplotlib.pyplot as plt #plotting to see the ratio
import seaborn as sns

sns.countplot(x=(clean['num']>0).map({False:'NO DISEASE', True:'DISEASE'}))
plt.show()
```

```python
plt.figure(figsize=(12,6)) #for heatmap all the data is converted to numeric datattype
cleann=clean.copy()
cleann=cleann.dropna()
cleann['thal']=cleann['thal'].map({'fixed defect':1,'normal':0,'reversable':2})
cleann['slope']=cleann['slope'].map({'downsloping':1,'flat':0,'upsloping':2})
cleann['restecg']=cleann['restecg'].map({'lv hypertrophy':1,'normal':0})
cleann['cp']=cleann['cp'].map({'typical angina':1,'asymptomatic':0,'non-anginal':2,'atypical angina':3})
cleann['dataset']=cleann['dataset'].map({'Cleveland':1,'Hungary':0,'Switzerland':2,'VA Long Beach':3})
cleann['fbs']=cleann['fbs'].astype(int)
cleann['sex']=cleann['sex'].map({'Male':1,'Female':0})
cleann=cleann.dropna()
sns.heatmap(data=cleann.corr(),annot=True) #heatmap through correlation
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression #importing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
cleann=cleann.dropna() 
print(cleann.shape)
cleann=cleann.select_dtypes(include=['int64','float64']) #only include numeric values
cleann['num']=cleann['num'].apply(lambda x:1 if x>0 else 0) #classifying all the disease categories as 1
X=cleann.drop(['id','num','dataset'],axis=1) #stating x and y
print(X.shape)
Y=cleann['num'].apply(lambda x:1 if x>0 else 0)
print(Y.shape)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25,random_state=42) #test and train
scaler=StandardScaler() #standardizing so that the data is confined between 0 and 1
```

```python
xtrain=scaler.fit_transform(xtrain) #transforming the data
xtest=scaler.transform(xtest)
```

```python
model=LogisticRegression(C=0.5,max_iter=10000,solver='liblinear',class_weight={0:1,1:2}) #applying logistic regression
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score #accuracy
acc=accuracy_score(ytest,ypred)
print(acc)
```

```python
from sklearn.metrics import confusion_matrix #confusion matrix
cm=confusion_matrix(ytest,ypred)
plt.figure(figsize=(6,3))
sns.heatmap(cm,annot=True)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred)) #classification report
```

```python
from sklearn.metrics import roc_curve,auc
#roc and auc
yprobs=model.predict_proba(xtest)[:,1]
fpr,tpr,threshold=roc_curve(ytest,yprobs)

plt.figure(figsize=(6,3))
roc_auc=auc(fpr,tpr)
plt.plot(fpr,tpr)
plt.plot([0,1],[1,0])
```

```python
features=X.columns #calculating importances
coefficients=model.coef_[0]

importances=pd.DataFrame(
    {'Feature':features,
    'importance':coefficients}
)

importances=importances.sort_values(by='importance',ascending=False)
print(importances)
```

```python
plt.figure(figsize=(12,6)) #plot
sns.barplot(x='feature',y='importance',data=importances)
plt.grid(True)
plt.show()
```
