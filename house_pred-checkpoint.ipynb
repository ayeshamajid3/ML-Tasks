```python
model=df[['price','area','stories','bedrooms','parking','bathrooms']] #picking the selective req columns 
model.head()
```

```python
x=model.drop('price',axis=1) #choosing x and y
y=model['price']
```

```python
from sklearn.linear_model import LinearRegression  #importing libs
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=42) #training

modell=LinearRegression()#regression
modell.fit(xtrain,ytrain)
ypred=modell.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error #checking the MAS,R2,MSE
mean=mean_absolute_error(ytest,ypred)
print(mean)
squ=mean_squared_error(ytest,ypred)
print(squ)
rr=r2_score(ytest,ypred)
print(rr)
```

```python
import matplotlib.pyplot as plt #plotting
import seaborn as sns
plt.figure(figsize=(12,6)) 
sns.scatterplot(x=ytest,y=ypred)
plt.plot([ytest.min(),ytest.max()],[ytest.min(),ytest.max()]) #stating starting and ending points for the line 
plt.show()
```
