```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=df)
plt.title('feature sepal length vs width')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

```python
sns.histplot(data=df,x='petal_length',kde=False,bins=10)
plt.xlabel('length')
plt.ylabel('frequency')
plt.show()
```

```python
sns.boxplot(x='sepal_length',y='sepal_width',data=df)
plt.show()
```
